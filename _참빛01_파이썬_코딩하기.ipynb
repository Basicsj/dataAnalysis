#파이썬 문법(Python Syntax)

### **string (문자열)**
 - 한 개 이상의 문자를 순서대로 나열 한 것
 - 문자열은 '(작은따옴표) 혹은 "(큰따옴표) 사이에 문자를 넣어서 생성
 - 문자열 자체에 ' , " 가 있는 경우에는 각각 그 반대의 기호로 생성

```python
a = '"Hello" World'
b = "I'm a student"
c = "Hello World.'

d = It's wonderful day"

print(a)
print(b)
print(c)
print(d)
```

#### **''' '''사용하여 표현 가능**
  - 차이점
    - '', "" -> 한 줄 문자열 표현
    - ''' ''' -> 여러 줄에 걸쳐 문자열 표현 가능

```python
c = '''
 Hello

 World
'''

d = """Hello
World"""

print(c)
print(d)
```

#### escape string (이스케이프 문자)
* 문자열내의 일부 문자의 의미를 달리하여 `특정한 효과`를 주는 것
* `\n` : new line
* `\t` : tab
image.png

```python
print('\nHello World\n\n')
print('***\t***\t***\t***')
```

#### **indexing & slicing string (문자열 인덱스 및 추출)**
  - 문자열의 각 문자는 순서가 있음
  - 이때 각 문자열의 순서를 **인덱스** 라고 함
  - 첫번째 문자부터 마지막까지 차례대로의 순서를 가짐
  - 첫번째 시작문자의 순서는 0으로 시작 (1이 아님)

```python
a = 'abcdefghijklmn'
len(a) # 길이
#print(a[13])
```

#### ** 인덱스 '-1' **
  - 다른 언어와는 달리, python의 경우 음수 인덱스를 지원
  - -1이 가장 마지막 인덱스를, -2가 마지막에서 두번째 인덱스를 의미...

```python
print(a[0])
print(a[1])

print(a[-1])
print(a[-2])
```

#### **문자열 slicing**
  - 인덱스가 하나의 문자만을 추출한다면,
  - slicing은 부분 문자열을 추출한다고 볼 수 있음
  - [시작:끝]와 같이 명시하여 (시작, 끝-1)에 해당하는 부분 문자열을 추출
  - 시작, 끝 인덱스가 생략이 되어 있다면, 0부터 혹은 끝까지로 간주

```python
a = '123456789012'
print("문자열의 길이는:", len(a))
print(a[0:11])
print(a[0:1])

print(a[:5])
print(a[3:])

print(a[:])
```

#### **문자열 함수**
  - 문자열은 여러가지 기능 제공을 위한 함수를 내장
  - 함수란 특정 기능을 하는 코드로 언제든지 호출하여 해당 기능을 사용 가능하도록 구성한 코드
  - 추후에 함수에 대해 자세히 다룰 예정

```python
a = 'abcdefg hijklmn'
a.upper()
```

* **replace**
 - 문자열 내의 특정 문자를 치환

```python
a = 'hello world'
a.replace('h', 'H')
```

* **split**
 - 문자열을 특정한 문자 구분하여(delimiter) 문자열의 리스트로 치환

```python
a = '2021/07/25'
a.split('/')
```

```python
a = 'hello world, what a nice weather'
a.split(' ')
```

* **format**
 - 문자열내의 특정한 값을 변수로부터 초기화하여 동적으로 문자열을 생성

```python
temperature = 32
prob = 70

a = '오늘 기온{}도 이고, 비올 확률은 {}% 입니다'.format(temperature, prob)
print(a)
```
