### 문자열에서 자주 사용하는 함수

```python
a = "Life is too short!"
print(f"문자(열) 개수 세기: {a.count('o')}")
print(f"문자(열) 위치: {a.find('f')}") # 없으면 '-1' 반환
print(f"문자(열) 위치: {a.index('f')}") # 없으면 오류
b = '*'
print(f"문자열 삽입: {b.join(a)}") # a에 b를 삽입
print(f"문자열 변환: {a.replace('Life', 'Your leg')}") # 문자열 대체
print(f"문자열 나누기: {a.split()}") # 공백으로 분리해서 리스트형으로 변환
print(f"문자열 나누기: {a.split('is')}") # is를 기준으로 나누므로 공백도 들어감
```

```python
# 관계 연산자(비교 연산자)
a = 100
print( a > 200)
print( a == 100)
print( a != 100)
```

```python
# 논리 연산자
a = 100
print(a > 100 and a==100)
print(a > 100 or a==100)
print( not( a==100) )
```

```python
# 멤버쉽 연산자
fruits = ['사과', '포도', '오렌지', '딸기']
print('딸기' in fruits)
print('키위' not in fruits)
print('키위' in fruits)

if '딸기' in fruits:
    print('딸기는 fruits 안에 있습니다.')
else:
    print('딸기는 fruits 안에 없습니다.')
```

```python
# for 반복문
for fruit in fruits:
    print(f'내가 좋아하는 과일 "{fruit}"')
```

### [실습문제] 동전 교환 프로그램 만들기

```python
a = int(input('동전으로 교환할 금액을 입력하세요: '))
b = a%500
c = b%100
d = c%50
print('-'*50)
print(f'500원짜리 ==> {a//500}개')
print(f'100원짜리 ==> {b//100}개')
print(f'50원짜리 ==> {c//50}개')
print(f'10원짜리 ==> {d//10}개')
print(f'잔돈 ==> {d%10}원')
```

```python
money = int(input('동전으로 교환할 금액을 입력하세요: '))
print('-'*50)
print(f'500원짜리 ==> {money//500}개')
money %= 500
print(f'100원짜리 ==> {money//100}개')
money %= 100
print(f'50원짜리 ==> {money//50}개')
money %= 50
print(f'10원짜리 ==> {money//10}개')
money %= 10
print(f'잔돈 ==> {money}원')
```
