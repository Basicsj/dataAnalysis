```python
 ## 박근혜쌤 미니프로젝트

Open In Colab

!pip install gTTS

파일에 추가하는 거 포함

from gtts import gTTS
from IPython.display import Audio, display
import time

def speaker(stmt):
    kor_wav = gTTS(stmt, lang='ko')
    kor_wav.save('kor.wav')
    display(Audio('kor.wav', autoplay=True))

# 파일에 내용을 추가하는 함수
def write_to_file(stmt):
    with open('output.txt', 'a') as file:  # 'a' 모드는 파일에 내용을 추가합니다.
        file.write(stmt + '\n')

#1. 단어등록
def w_add(dictionary):
    while True:
        key = input('새 단어 : ')
        if key in dictionary:
            msg = f'입력하신 {key}는 이미 등록되어 있는 단어입니다.'
            print(msg)
            speaker(msg)
            write_to_file(msg)
        else:
            break
    value = input('의 뜻 : ')
    msg = f'등록단어: {key}, 뜻: {value}'
    print(msg)
    speaker(f'새로 등록하신 단어는 {key}이고, 뜻은 {value}입니다.')
    write_to_file(msg)  # 등록된 단어를 파일에 저장
    dictionary[key] = value

#2. 단어검색
def w_search(dictionary):
    search_term = input('검색할 단어 또는 뜻 : ')
    found = False

    # key 검색
    if search_term in dictionary:
        meaning = dictionary[search_term]
        msg = f'검색하신 단어는 {search_term}이고 뜻은 {meaning}입니다.'
        print(msg)
        speaker(msg)
        write_to_file(msg)  # 검색된 단어를 파일에 저장
        found = True

    # value 검색
    value_searched = False  # value 검색 여부를 나타내는 변수 추가
    for key, value in dictionary.items():
        if value == search_term:
            msg = f'검색하신 뜻은 {value}이고 단어는 {key}입니다.'
            print(msg)
            speaker(msg)
            write_to_file(msg)  # 검색된 단어를 파일에 저장
            found = True
            value_searched = True  # value 검색이 실행된 경우 변수를 True로 설정

    # key 검색이나 value 검색이 모두 실행되지 않았고 value 검색이 실행된 적도 없는 경우에만 일부 검색 수행
    if not found and not value_searched:
        for key, value in dictionary.items():
            if search_term in key or search_term in value:
                msg = f'검색하신 일부를 포함하는 단어는 {key}이고 뜻은 {value}입니다.'
                print(msg)
                speaker(msg)
                write_to_file(msg)  # 검색된 단어를 파일에 저장
                time.sleep(5)
                found = True

    # 검색 결과가 없는 경우
    if not found:
        msg = f'"{search_term}"에 대한 검색 결과가 없습니다.'
        print(msg)
        speaker(msg)
        write_to_file(msg)  # 검색결과가 없는 단어를 파일에 저장

#3. 단어삭제
def w_delete(dictionary):
    key = input('삭제할 단어 : ')
    if key in dictionary:  # 딕셔너리에 입력된 키 값이 존재하는지 확인
        del dictionary[key]  # 키 값이 존재하면 해당 단어를 삭제
        msg = f'"{key}" 단어 삭제됨!'
        print(msg)
        speaker(f'삭제하신 단어는 {key}입니다.')
    else:
        msg = f'"{key}" 단어는 미등록 단어'
        print(msg)
        speaker(f'{key}는 미등록 단어입니다.')
    write_to_file(msg)

#4. 모든단어보기
def w_show(dictionary):
    msg = str(dictionary)
    print(msg)
    time.sleep(3)
    speaker(f'등록되어 있는 모든 단어는 {msg}입니다.')
    write_to_file(msg)

#0. 종료하기
def quit():
    msg = '프로그램을 종료합니다.'
    print(msg)
    speaker(msg)
    write_to_file(msg)

dictionary = dict()    # 빈 딕셔너리 생성

# 메뉴 출력 여부를 나타내는 변수
menu_displayed = False

while True:
    menu = '''---선택메뉴--
    1. 단어 등록
    2. 단어 검색
    3. 단어 삭제
    4. 모든 단어 보기
    0. 종료하기
    --------------------
    원하시는 메뉴를 선택해주세요'''
    print(menu, end='')

    time.sleep(5)

    if not menu_displayed:
        speaker(menu)
        menu_displayed = True

    time.sleep(2)

    select = input('선택:')

    if select == '1':
        speaker('새 단어를 등록하세요')
        w_add(dictionary)

    elif select == '2':
        speaker('단어를 검색하세요')
        w_search(dictionary)

    elif select == '3':
        speaker('삭제할 단어를 입력하세요')
        w_delete(dictionary)

    elif select == '4':
        speaker('모든 단어를 보여주세요')
        w_show(dictionary)

    elif select == '0':
        quit()
        break
```

```python
## 이윤진쌤 미니프로젝트


# 미니 프로젝트 만들기 : 지뢰찾기
# 3x3 보드에서 지뢰를 찾아보기

import random

class MineGame:
  def __init__(self):
    self.board = [str(i) for i in range(9)]
    self.mine_index = random.randint(0, 8)

  def create_board(self):
    for i in range(3):
      for row in [self.board[i*3:(i+1)*3]]:
        print("|" + " | ".join(row) + "|")

  def user(self):
    while True:
      try:
        user = int(input("보드에서 지뢰를 찾아주세요: "))
        if user not in [0, 1, 2, 3, 4, 5, 6, 7, 8]:
          print("잘못 입력하셨습니다. 0~8사이의 숫자를 입력해주세요.")
        elif self.board[user] == "X":
          print("이미 선택한 위치입니다. 다른위치에서 지뢰를 찾아주세요.")
          continue
        elif user == self.mine_index:
          self.board[user] = "*"
          print("지뢰를 찾았습니다.")
          return False
        else:
          self.board[user] = "X"
          print("지뢰를 찾지 못했습니다.\n다시 찾아주세요.")
          return True
      except ValueError:
        print("잘못된 입력입니다. 숫자를 입력해 주세요.")


  def show_board(self):
    for i in range(3):
      for row in [self.board[i*3:(i+1)*3]]:
        print("|" + " | ".join(row) + "|")

  def play_game(self):
    self.create_board()
    while True:
      self.user()
      self.show_board()
      if not self.user():
        break
    self.show_board()


MineGame().play_game()

# 수정해야 할 부분:
# 루프 오류 : 지뢰를 찾았을때 바로 종료되지 않음, 잘못된 값을 두번 입력하면 더이상 루프를 돌지않고 종료됨.
```

```python
## 이봉희쌤 미니프로젝트 visaul studio

from tkinter import *
from tkinter.filedialog import *
from tkinter.messagebox import askokcancel
from PIL import Image, ImageTk
import os

def func_open():
    global current_index, image_files, image_labels, directory, zoom_level, degree
    directory = askdirectory(parent=window, title='사진이 저장된 폴더를 선택하세요')
    if directory:
        image_files = [f for f in os.listdir(directory)
                       if f.endswith(('.jpg', '.jpeg','.JPG', '.bmp', '.png', '.tif', '.gif'))]
        if image_files:
            current_index = 0
            zoom_level = 0.5
            degree = 0
            display_image(directory, image_files[current_index])

def display_image(dir, filename):
    global current_index, image_files, image_labels, zoom_level, degree
    img_path = os.path.join(dir, filename)
    photo = Image.open(img_path)
    photo = photo.resize((int(photo.width*zoom_level), int(photo.height*zoom_level)))
    photo = photo.rotate(degree, expand=True)
    window.geometry(str(photo.width) + 'x' + str(photo.height))
    img = ImageTk.PhotoImage(photo)
    image_labels['current'].configure(image=img)
    image_labels['current'].image = img
    fnLabel.configure(text=filename)
    btnPrev.pack(side=LEFT)
    fnLabel.pack(side=LEFT)
    btnNext.pack(side=RIGHT)
    image_labels['current'].pack()

def clickPrev():
    global current_index, image_files, image_labels, directory, zoom_level
    if current_index > 0:
        current_index -= 1
        zoom_level = 0.5
        degree = 0
        display_image(directory, image_files[current_index])

def clickNext():
    global current_index, image_files, image_labels, directory, zoom_level
    if current_index < len(image_files) - 1:
        current_index += 1
        zoom_level = 0.5
        degree = 0
        display_image(directory, image_files[current_index])

def func_save():
    saveFp = asksaveasfile(parent=window, mode='w',defaultextension='.jpg',
                           filetypes=(('JPG 파일','*.jpg;*.jpeg'),('모든 파일', '*.*')))

def func_del():
    global current_index, image_files, image_labels, zoom_level
    if askokcancel("파일 삭제", "파일을 정말 삭제하시겠습니까?"):
        file_to_delete = os.path.join(directory, image_files[current_index])
        os.remove(file_to_delete)
        image_files.pop(current_index)
        if current_index >= len(image_files):
            current_index = len(image_files) - 1
        zoom_level = 0.5
        if image_files:
            display_image(directory, image_files[current_index])
        else:
            image_labels['current'].configure(image=None)
            fnLabel.configure(text="")

def func_zoomin(event=None):
    global zoom_level, image_files, current_index, directory
    zoom_level += 0.1
    display_image(directory, image_files[current_index])


def func_zoomout(event=None):
    global zoom_level, image_files, current_index, directory
    if zoom_level > 0:
        zoom_level -= 0.1
        display_image(directory, image_files[current_index])

def func_rotateLeft(event=None):
    global degree, image_files, current_index, directory
    degree += 90
    display_image(directory, image_files[current_index])

def func_rotateRight(event=None):
    global degree, image_files, current_index, directory
    degree -= 90
    display_image(directory, image_files[current_index])


def func_exit():
    window.quit()
    window.destroy()

window = Tk()
window.title('앨범 정리 프로그램')
window.geometry('500x500')

window.bind('<Up>', func_zoomin)
window.bind('<Down>', func_zoomout)
window.bind('<Left>', func_rotateLeft)
window.bind('<Right>', func_rotateRight)

mainMenu = Menu(window)
window.config(menu=mainMenu)

upFrame = Frame(window)
upFrame.pack()
downFrame = Frame(window)
downFrame.pack()

fileMenu = Menu(mainMenu)
mainMenu.add_cascade(label='파일', menu=fileMenu)
fileMenu.add_command(label='열기', command=func_open)
fileMenu.add_command(label='저장', command=func_save)
fileMenu.add_command(label='삭제', command=func_del)
fileMenu.add_separator()
fileMenu.add_command(label='종료', command=func_exit)

btnPrev = Button(upFrame, text='<<이전', command=clickPrev)
btnNext = Button(upFrame, text='다음>>', command=clickNext)

fnLabel = Label(upFrame)
image_labels = {'current': Label(downFrame)}
frontImage = PhotoImage(file='./images/family-album.gif')
image_labels['current'].configure(image=frontImage)
image_labels['current'].pack()

window.mainloop()
```

```python
## 김윤이쌤 미니프로젝

# 파이숑과 함께하는 UP & Down 게임
import random

desc = f'''[[ 파이숑과 함께하는 UP & Down 게임! ]] {"ㅛ"*13}
1. 파이숑이 선택할 숫자의 자릿수를 지정해주세요.
   ( 1 => 1 ~ 9 / 2 => 10 ~ 99 / 3 => 100 ~ 999 )
2. 도전 값에 따라 UP & Down 을 알려줍니다.
3. 도전 실패 5회 마다 결정적 힌트를 얻을 수 있어요.
4. 도전 시작 후 그만하기를 원하는 경우 입력창에 '0'을 넣어주세요.
5. 과연 몇번만에 파이숑의 숫자를 맞출 수 있을까요^^
{"ㅠ"*33}
'''
print(desc)

pythongNo, tryCnt = 0, 0  # 파이숑 선택 숫자, 시도 횟수

# >>>> 파이숑에게 선택할 숫자의 자릿수 받기:
digit = int(input('파이숑이 선택할 숫자의 자릿수를 입력해주세요: '))

# >>> 파이숑 숫자 선택(random)
pythongNo = random.randint(int(f'1{"0"* (digit-1) }'),int(f'{"9"* digit}'))
print(f'{"-"*66}\n>>> 파이숑 숫자 1{"0"* (digit-1) } ~ {"9"* digit} 중 선택 완료!. ({("■"*digit)})\n>>> 도전 시작~')

while True:

    tryCnt += 1

    # >>> 답변 입력
    answer = int(input(f'{"-"*50}\n파이숑이 선택한 {digit}자리의 숫자는 무엇일까요?: '))
    print('-'*66)

    # >>> 그만하기(종료)
    if answer == 0:
        tyrGo = int(input('>>> 게임을 중단하시겠습니까?(예: 1 / 아니오: 2 ): '))

        if tyrGo == 1:
            print(f'\n>>> (●´⌓`●) 파이숑이 선택한 숫자는 ( \033[91m {pythongNo} \033[30m ) 였어요.')
            # # \033[(숫자)m : 원하는 색에 대해 ANSI 표준으로 정해진 숫자 지정
            break

        else:
            continue

    # >>> 정답
    if pythongNo == answer:
        print(f'\n>>> \033[91m( {pythongNo} )정답!! ✌(❛‿❛)✌ {tryCnt}회 만에 성공했어요~')
        break

    # >>> UP & Down
    if pythongNo > answer:
        print(f'>>> \033[94m▼Down▼\033[30m 숫자를 좀 더 올려 볼까요~')
    else:
        print(f'>>> \033[95m▲UP▲\033[30m 숫자를 좀 더 내려 보아요~ ')


    # >>> 힌트
    if tryCnt % 5 == 0:
        hintDigit = int(input(f'\n>>> [⭐힌트] 몇번째 숫자를 알려드릴까요?: '))

        hint = ''
        for i in range(digit):
            if i+1 == hintDigit:
                hint += str(pythongNo)[hintDigit-1:hintDigit]
            else:
                hint += '■'

        print(f'>>> 파이숑이 선택한 숫자는 {hint} 입니다\n')
```

```python
## 김승미쌤 미니프로젝트


# 주소록 프로그램 만들기

menu = ''' ---------[선택메뉴]--------
1. 연락처 입력
2. 연락처 검색
3. 연락처 삭제
4. 연락처 보기
0. 종료하기
'''

address_book = {}  # 주소록을 저장할 딕셔너리

# 함수 선언부분
def book_add():
    name = input('이름:')
    phone_number = input('전화번호:')
    email  = input('E-mail:')
    addr =input('Address:')

    # 주소록에 딕셔너리 형태로 저장 {이름 key, value 값으로 전화번호, 이메일, 주소를 딕셔너리에 저장}
    address_book[name] = {'전화번호': phone_number,'E-mail': email,'Address':addr}
    print(f"'{name}'연락처가 등록되었습니다.")

def book_search():   # 연락처 검색
    name_search = input('연락처 검색: ')
    if name_search in address_book:                 # 현재 등록된 딕셔너리에  이름(key)가 있는지 검색
        contact_info = address_book[name_search]    # 해당 이름에 대한 연락처(value) 정보 가져오기
        print(f"이름: {name_search}")                # 주소록 정보 출력
        print(f"전화번호: {contact_info['전화번호']}")
        print(f"E-mail: {contact_info['E-mail']}")
        print(f"Address: {contact_info['Address']}")
    else:
        print("연락처가 등록되어 있지 않습니다..")
        input_name = input('새로운 연락처를 등록하시겠습니니까?(등록:Y, 미등록:N):')  # 미등록된 연락처 추가할지 여부 확인
        if input_name =='Y' or 'y':
            book_add()  # 연락처 추가 함수 호출
        else:
            print(f'등록하지 않습니다.')

def book_delete():  # 연락처 삭제
    name_Delete = input('삭제할 연락처 :')     # 삭제할 연락처 이름 받기
    if name_Delete in address_book:           # 삭제할 연락처가 딕셔너리에 있는지 검색
        del address_book[name_Delete]         # 딕셔너리 주소록에서 연락처 삭제
        print(f"연락처가 삭제 되었습니다: {name_Delete}")
    else:
        print(f"등록되지 않은 이름입니다..")

def book_show():
    for name, contact_info in address_book.items():   # 주소록의 모든 연락처에 대해 반복
        print(f"이름: {name}")                        # 현재 등록된 모든 {이름 key, value 값으로 전화번호, 이메일, 주소} 출력
        print(f"전화번호: {contact_info['전화번호']}")
        print(f"E-mail: {contact_info['E-mail']}")
        print(f"Address: {contact_info['Address']}")


# 메인코드 부분
while True:
    print(menu)
    select = input('-->메뉴 선택: ')
    if select == '0':    # 프로그램 종료
        break
    elif select == '1':  # 연락처 입력
        book_add()       # 연락처 입력 함수 호출
    elif select == '2':  # 연락처 검색
        book_search()    # 연락처 검색 함수 호출
    elif select == '3':  # 연락처 삭제
        book_delete()    # 연락처 삭제 함수 호출
    elif select == '4':  # 연락처 보기
        book_show()      # 연락처 보기 함수 호출

print('프로그램을 종료합니다.')
```
