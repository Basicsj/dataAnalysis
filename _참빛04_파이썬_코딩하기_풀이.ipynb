## while :

### 반복문을 사용하여 0~ 9 까지 출력하기
image.png

````python
n = 0
while n < 1000 :
    print(n)
    n = n+1
````

### 100 ~ 0 까지 5의 배수 출력하기
*  초기값 n = 100
* n > 0 까지 반복하기
* n의 값을 5씩 빼기
* 출력하기



````python
n=100
while n>0:
    print(n)
    n -= 5 # n = n-5

print("마지막 n의값은", n)
````

###  1~ 100 사이의 3의 배수 출력하기

````python
a = 1
while a <= 100:
    if a % 3 == 0:
        print(a)
    else:
        pass
    a += 1
````

## for :

````python
for a in range(100):   # 0부터 99까지의 범위를 뜻함
    print(a)
````
